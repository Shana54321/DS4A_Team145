```python
from pytrends.request import TrendReq
import csv
import pandas as pd
import numpy
import time

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
keywords = pytrends.suggestions(keyword='pepsi')
df = pd.DataFrame(keywords)
df = df.drop(columns='mid')   # This column makes no sense
pepsi_list = list(df['title'])
pytrends.build_payload(pepsi_list, cat=0, timeframe='2005-01-01 2021-06-01', geo='', gprop='')
pepsi_interest_over_time_df = pytrends.interest_over_time()
pepsi_df = pepsi_interest_over_time_df.drop(columns=['PepsiCo', 'Sabyasachi Mukherjee', 'Pepsin', 'The Pepsi Bottling Group'])
pepsi_df = pepsi_df.drop(columns="isPartial")
pepsi_df
```
