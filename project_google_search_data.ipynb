```python
from pytrends.request import TrendReq
import csv
import pandas as pd
import numpy
import time
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import pingouin

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
keywords = pytrends.suggestions(keyword='pepsi')#gets google trends to give a list of suggested search terms for Pepsi
df = pd.DataFrame(keywords) 
df = df.drop(columns='mid')   # This column makes no sense
pepsi_list = list(df['title'])# Converts the dataframe to a list of pepsi alternate names
pytrends.build_payload(pepsi_list, cat=0, timeframe='2005-01-01 2021-06-01', geo='', gprop='') #runs the list of alternate Pepsi names to display the google numders for each word; also allows a timeperiod for searches 
pepsi_interest_over_time_df = pytrends.interest_over_time()
pepsi_df = pepsi_interest_over_time_df.drop(columns=['PepsiCo', 'Sabyasachi Mukherjee', 'Pepsin', 'The Pepsi Bottling Group'])#Drops the colums with name suggestions that return no data
pepsi_df = pepsi_df.drop(columns="isPartial")#Drops the isspartial column that gives number where the name is partially a part of the search
pepsi_df
```
